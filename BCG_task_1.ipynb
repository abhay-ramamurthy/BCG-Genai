```python
import pandas as pd
import numpy as np
df = pd.read_csv('Abhay54.csv')
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df
```

```python


df.fillna(0, inplace=True)
df
```

```python


# Calculate year-by-year growth rates
df['Total Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Total Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operating Activities Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
result = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Total Assets Growth (%)': 'mean',
    'Total Liabilities Growth (%)': 'mean',
    'Cash Flow from Operating Activities Growth (%)': 'mean'
}).reset_index()
result
```

Summary of findings --
 Apple - apple had an negative average year on year growth for the past 3 year's. it has experienced decrease (negative growth) in revenue growth, net income, total assets. but on the positive side of things - the mean rate of growth of liabilities has also decreased.
 The same is the case of Tesla and Microsoft as well but in different proportions as seen in the above growth figures


```python
result.to_csv('result.csv', index=False)
```
